# Spatial Resolution: Sub and Over-Sampling

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
img=data.camera()
plt.imshow(img)
Image.fromarray(img).save('camera.jpg')
```

# Sub-Sampling
Means resizing to smaller size image

```python
height, width=img.shape
th=8
height=int(height/th)
width=int(width/th)
imgNew=np.ndarray((height, width), dtype=np.uint8)
for i in range(0,height):
    for j in range(0,width):
        imgNew[i][j]=img[i*th][j*th]
plt.imshow(imgNew)
Image.fromarray(imgNew).save('cameraSub.jpg')
```

# Over-Sampling:
Resizing image to larger size

```python
h,w=imgNew.shape
th=8
h=h*th
w=w*th
imgNew2=np.ndarray((h,w),dtype=np.uint8)
for i in range(0,h):
    for j in range(0,w):
        imgNew2[i][j]=imgNew[int(i/th)][int(j/th)]
plt.imshow(imgNew2)
imfile=Image.fromarray(imgNew2)
imfile.save('cameraOver.jpg')
# print(imgNew2.shape)
```

## Log Intensity Transformation

```python
import math
img=data.grass()
height, width=img.shape
for i in range(0,height):
    for j in range(0,width):
        img[i][j]=math.log10(1+img[i][j])
plt.imshow(img, 'gray')
```

## Power-Law (Gamma) Transformation

```python
c=1
img=data.astronaut()
# height, width=img.shape
# for i in range(0,height):
#     for j in range(0,width):
img=np.array(c*img**0.7, dtype='uint8')
plt.imshow(img)
Image.fromarray(img).save('Temp.jpg')
```
